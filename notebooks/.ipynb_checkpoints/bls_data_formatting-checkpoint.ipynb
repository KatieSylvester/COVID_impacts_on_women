### Read-in Bureau and Labor Statistics on overall population:
- general population
- civilian labor force
- employment level

### Get data ready to merge into one dataframe


##### NOTE TO SELF:  Before reading in BLS charts, add zeroes to the missing 2021 data in the CSV to avoid data type errors. Then, don't forget later to drop those rows (April-Dec 2021)

```python
civilian_pop=pd.read_csv('../data/civilian_noninstitutional_population_16plus_LNU00000000r.csv', dtype = 'Int64')
civilian_labor_force=pd.read_csv('../data/civilian_labor_force_level_16plus_LNU01000000.csv')
employment_level=pd.read_csv('../data/employment_level_16plus_LNU02000000r.csv')
```

```python
# creates a numeric month column

def month_number(counter):
    if (counter['month_name'] == 'Jan'):
        return 1
    elif (counter['month_name'] == 'Feb'):
        return 2
    elif (counter['month_name'] == 'Mar'):
        return 3
    elif (counter['month_name'] == 'Apr'):
        return 4
    elif (counter['month_name'] == 'May'):
        return 5
    elif (counter['month_name'] == 'Jun'):
        return 6
    elif (counter['month_name'] == 'Jul'):
        return 7
    elif (counter['month_name'] == 'Aug'):
        return 8
    elif (counter['month_name'] == 'Sep'):
        return 9
    elif (counter['month_name'] == 'Oct'):
        return 10
    elif (counter['month_name'] == 'Nov'):
        return 11
    elif (counter['month_name'] == 'Dec'):
        return 12
    else:
        return 'adjust'
```

```python
# melts the civilian population chart to one column 
# and sets up the dataframe which we will add all of the other metrics to it

civilian_pop_tall = civilian_pop.melt(id_vars = ['Year'], value_vars=['Jan', 'Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], var_name=['month_name'], value_name='pop')
civilian_pop_tall=civilian_pop_tall.rename(columns={'Year':'year',  'pop':'civilian_pop'})
civilian_pop_tall["month"] = civilian_pop_tall.apply(month_number, axis=1)
civilian_pop_tall['date'] = pd.to_datetime(civilian_pop_tall[['year', 'month']].assign(day=1))
civilian_pop_tall['civilian_pop'] = civilian_pop_tall['civilian_pop'].astype('Int64')
#only needed for first civilian population file which will be used as the LEFT df in the 1st join
civilian_pop_tall=civilian_pop_tall[['year','month_name','month','date','civilian_pop']]

civilian_pop_tall = civilian_pop_tall.sort_values('date')
civilian_pop_tall.info()
```

```python
# defines a function to melt each of the Bureau of Labor Statistics charts of each metric
# so I can merge them together into one dataframe to analyze

def melt_df(df, df_string):
    df= df.melt(id_vars = ['Year'], value_vars=['Jan', 'Feb', 'Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], var_name=['month_name'], value_name=df_string)
    df = df.rename(columns={'Year':'year'})
    df["month"] = df.apply(month_number, axis=1)
    df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
    df = df.sort_values('date')
    return df
```

#### Melting each overall population BLS chart (each metric is contained in a separate chart)

```python
employment_level_tall = melt_df(employment_level,'employment_level')
employment_level_tall.head()
```

```python
labor_force_tall = melt_df(civilian_labor_force,'total_labor_force')
labor_force_tall.head()
```

### Read-in and format Women 20+ Statistics

```python
women_pop=pd.read_csv('../data/civilian_noninst_pop_women_20plus_LNU00000026r.csv')
women_lf_level=pd.read_csv('../data/civilian_labor_force_level_women_20plus_LNU01000026r.csv')
women_ee_level=pd.read_csv('../data/employment_level_women_20plus_LNU02000026r.csv')
women_unempl_level=pd.read_csv('../data/unemployment_level_women_20plus_LNU03000026r.csv')
```

```python
women_pop_tall = melt_df(women_pop, 'women_pop')
women_lf_level_tall = melt_df(women_lf_level,'women_lf_level')
women_ee_level_tall = melt_df(women_ee_level,'women_ee_level')
women_unempl_level_tall = melt_df(women_unempl_level, 'women_unempl_level')
```

#### Merging data

```python
data_combined = pd.merge(civilian_pop_tall,employment_level_tall[['date','employment_level']],on='date', how='left')
```

```python
data_combined = pd.merge(data_combined,labor_force_tall[['date','total_labor_force']],on='date', how='left')
```

```python
data_combined['overall_unemployment_rate']=round((100*(data_combined['total_labor_force']-data_combined['employment_level'])/data_combined['total_labor_force']),1)
```

```python
data_combined.head()
```

```python
data_combined = pd.merge(data_combined,women_pop_tall[['date','women_pop']],on='date', how='left')
```

```python
data_combined = pd.merge(data_combined,women_lf_level_tall[['date','women_lf_level']],on='date', how='left')
```

```python
data_combined = pd.merge(data_combined,women_ee_level_tall[['date','women_ee_level']],on='date', how='left')
```

```python
# calculation of women in the labor force participation rate
data_combined['women_lf_part_rate']=round(100*(data_combined['women_lf_level']/data_combined['women_pop']),1)
```

```python
######
```

```python
# calculation of women not in the labor force
data_combined['women_not_in_labor_force']=(data_combined['women_pop'] - data_combined['women_lf_level'])
```

```python
data_combined = pd.merge(data_combined,women_unempl_level_tall[['date','women_unempl_level']],on='date', how='left')
```

```python
# calculation of women's unemployment rate
data_combined['womens_unempl_rate'] = round((100*data_combined['women_unempl_level']/(data_combined['women_lf_level'])),1)
```
