```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
supersectors=pd.read_csv('../data/SuperSectorJobLevels.csv')
```

```python
supersectors.head()
```

```python
supersectors['% Women Field'] = supersectors['Feb2020_women']/supersectors['Feb2020_total']
```

```python
supersectors['12MO Net Change All_Neg']=supersectors['Feb2021_total']-supersectors['Feb2020_total']
```

```python
supersectors['12MO Net Change_Neg']=supersectors['Feb2021_women']-supersectors['Feb2020_women']
```

```python
supersectors['% Women Job Loss'] = supersectors['12MO Net Change_Neg']/supersectors['12MO Net Change All_Neg']
```

```python
supersectors['12Mo % Changeall']=supersectors['12MO Net Change All_Neg']/supersectors['Feb2020_total']
```

```python
supersectors['12Mo % Change']=supersectors['12MO Net Change_Neg']/supersectors['Feb2020_women']
```

```python
supersectors['12Mo Net Change']=-1000*supersectors['12MO Net Change_Neg']
```

```python
supersectors['12Mo Net Change All']=-1000*supersectors['12MO Net Change All_Neg']
```

```python
supersectors['% Men Field']=1-supersectors['% Women Field']
```

```python
supersectors['% Men Job Loss']=1-supersectors['% Women Job Loss']
```
